# **Fundamentals Perceptron**

```python
import numpy as np
```

```python
x = np.array([[1, 2.1, 3.9]])
w = np.array([[0.5, 0.5, 0.5]]).reshape(-1, 1)

np.dot(x, w)
```

```python
identity = lambda x: x
binary = lambda x: 1 if x >= 0 else 0
sigmoid = lambda x: 1 / (1 + np.exp(-x))
relu = lambda x: x if x > 0 else 0


def sim_perceptron(input: np.array, weights: np.array, bias_w: float, activation_fn: callable) -> float:
    """ Simulate a perceptron with the given input, weights, bias and activation function.
    
    Args:
        input (np.array): Numpy array of input values
        weights (np.array): Numpy array of weights
        bias_w (float): Bias weight
        activation_fn (callable): Activation function
    
    Returns:
        float: Output of the perceptron
    """
    z = np.dot(input, weights) + bias_w
    return activation_fn(z)

sim_perceptron(x, w, 0.3, relu)
```
