## **Math Fundamentals**

### **Norm of a Vector**

- The norm of a vector is also known as the *Euclidean norm*, *2-norm*, *magnitude*, or *length* of the vector. It quantifies the "size" or "length" of the vector in a geometric sense.
- Mathematically, the norm of a vector $ v $ in $ n $-dimensional space is defined as the square root of the dot product of the vector with itself. It can be expressed as:
  $$
  ||v|| = \sqrt{\sum_{i=1}^{n} v_i^2} = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}
  $$
  where $ v_1, v_2, \dots, v_n $ are the components of the vector $ v $.
- Geometrically, the norm represents the Euclidean distance of the vector from the origin $ (0, 0, \dots, 0) $ in $ n $-dimensional space.
- For vectors in 2D or 3D space, this can be visualized as the straight-line distance from the origin to the point represented by the vector.

### **Normalization of a Vector**

- A *normalized vector* is often referred to as a *unit vector*, which is a vector with a length (or norm) of 1.
- Normalizing a vector is the process of scaling the vector so that its norm becomes 1, while maintaining its direction.
- Mathematically, the normalized vector $ \hat{v} $ is defined as the original vector $ v $ divided by its norm $ ||v|| $:
  $$
  \hat{v} = \frac{v}{||v||}
  $$
  where $ ||v|| $ is the norm of the vector $ v $.

By normalizing a vector, we retain its direction but scale its magnitude to 1. This is particularly useful in many applications, such as in machine learning and physics, where only the direction of a vector matters and not its scale. 

##### **Additional Notes**:
- For a vector $ v $ the result $ \hat{v} $ is always a unit vector, meaning $ ||\hat{v}|| = 1 $
- The process of normalization is essential when comparing vectors based on direction alone, as it removes the influence of their magnitudes.
- If the norm of the vector $ v $ is zero (i.e., $ v $ is the zero vector), normalization is undefined because division by zero is not possible.

### **Dot Product**

- The **dot product** (also known as the *scalar product* or *inner product* in more general terms) is an operation that combines two vectors and returns a scalar value.
- **The length (or norm) of the vectors matters** when computing the dot product, as both the magnitude and the angle between the vectors influence the result.
- **Dot product with normalized vectors** is equivalent to the cosine of the angle between them, and it is often referred to as *cosine similarity* in many applications, especially in machine learning and information retrieval.

##### **Coordinate Definition**:
The dot product of two vectors $ v $ and $ w $ in $ n $ dimensional space is calculated as the sum of the products of their corresponding components:
$$
\text{dot product} = v \cdot w = \sum_{i=1}^{n} v_i w_i = v_1 w_1 + v_2 w_2 + \dots + v_n w_n
$$

##### **Geometric Definition**:
The dot product can also be defined geometrically as the product of the magnitudes (norms) of the two vectors and the cosine of the angle $ \theta $ between them:
$$
v \cdot w = ||v|| \, ||w|| \cdot \cos(\theta)
$$
Here, $ ||v|| $ and $ ||w|| $ are the magnitudes of vectors $ v $ and $ w $ and $ \theta $ is the angle between them.

##### **Interpretation**:
- **When the dot product is greater than 0**: The vectors are pointing in the same direction. 
    - For normalized vectors, the maximum value of the dot product is 1 (when the vectors are exactly aligned).
- **When the dot product is less than 0**: The vectors are pointing in opposite directions. 
    - For normalized vectors, the minimum value of the dot product is -1 (when the vectors are exactly opposite).
- **When the dot product is equal to 0**: The vectors are orthogonal (perpendicular to each other). This means the cosine of the angle between them is zero.

### **Cosine Similarity**

- **The length (magnitude) of the vectors does not matter** when calculating cosine similarity. Instead, it is purely based on the angle between the vectors.
- **Cosine similarity** is a measure of similarity between two non-zero vectors, and it is defined as the cosine of the angle $ \theta $ between them.
- The result of cosine similarity depends on the type of vectors:
    - For non-negative vectors, the result is between 0 and 1. Cosine similarity is often used in text analysis, in combination with techniques like [TF-IDF](https://en.wikipedia.org/wiki/Tf%E2%80%93idf), to assess document similarity.
    - For general vectors, the result can range from -1 to 1:
        - **-1** indicates no similarity (the vectors are diametrically opposite).
        - **1** indicates exact similarity (the vectors are perfectly aligned in the same direction).
        - **0** indicates orthogonality (no similarity, the vectors are at a 90-degree angle to each other).

##### **Formula**:
Cosine similarity is calculated as the **dot product** of two vectors $ v $ and $ w $ divided by the product of their **norms** (magnitudes):
$$
\text{cosine similarity} = \cos(\theta) = \frac{v \cdot w}{||v|| \, ||w||}
$$
This formula quantifies how similar the vectors are by considering both their direction and magnitude, but with a focus on the angle between them.

### **Cosine Distance**

- **The length (magnitude) of the vectors does not matter** when calculating cosine distance, just like cosine similarity.
- **Cosine distance** is a measure of dissimilarity between two non-zero vectors. It is derived from cosine similarity and quantifies how far apart two vectors are in terms of their orientation.
  The formula for cosine distance is:
  $$
  \text{cosine distance} = 1 - \text{cosine similarity}
  $$

Since cosine similarity ranges from -1 to 1 (or 0 to 1, depending on the vector types), cosine distance typically ranges from 0 to 2:
- **0** indicates that the vectors are perfectly aligned (i.e., cosine similarity is 1).
- **2** indicates that the vectors are completely opposite (i.e., cosine similarity is -1).
- **1** indicates orthogonality, where the vectors are perpendicular to each other (i.e., cosine similarity is 0).

---

### **Euclidean Distance**

- **The length (magnitude) of the vectors matters** when calculating Euclidean distance, as it measures the straight-line distance between two points in space.
- **Euclidean distance** is the length of the line segment connecting two points, represented by the vectors $ v $ and $ w $ in Euclidean space (which is the typical "flat" space we're familiar with, such as 2D or 3D space).
  
  The formula for Euclidean distance between two vectors $ v $ and $ w $ is:
  $$
  \text{euclidean distance} = \sqrt{\sum_{i=1}^{n} (v_i - w_i)^2} = \sqrt{(v_1 - w_1)^2 + (v_2 - w_2)^2 + \dots + (v_n - w_n)^2} = ||v - w||
  $$
  This formula calculates the straight-line distance between the vectors $ v $ and $ w $ by summing the squared differences of their corresponding components and then taking the square root of that sum.

##### **Additional Notes**:
- **Geometrically**, the Euclidean distance represents the direct, "as-the-crow-flies" distance between the two points, which is the shortest possible path in Euclidean space.
- Euclidean distance is sensitive to the magnitude of the vectors, which means that two vectors pointing in the same direction but with very different lengths will have a large Euclidean distance.

## **Math Fundamentals using Python**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as la
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances, euclidean_distances
from ipywidgets import *
```

```python
def plot_2d_vectors(*args, legend_labels=[], legend_loc=0, colormap='tab10', title="Vectors", xlabel="x", ylabel="y"):
    cmap = plt.get_cmap(colormap, len(args))
    for i, vec in enumerate(args):
        try:
            label = legend_labels[i]
        except IndexError:
            label = f"Vector {i+1}"
        plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=cmap(i), label=label)
    
    max_val = np.ceil(np.max([*args]) + 1)
    min_val = np.floor(np.min([*args]) - 1 if np.min([*args]) < 0 else 0)
    
    plt.xlim(min_val, max_val)
    plt.ylim(min_val, max_val)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc=legend_loc)
    
    ax = plt.gca()
    ax.set_aspect('equal')
    ax.set_xticks(np.arange(min_val, max_val+1))
    ax.set_yticks(np.arange(min_val, max_val+1))
    
    plt.grid()
```

```python
plot_2d_vectors(np.array([0, 2]), np.array([1, 2]), np.array([2, 2]), np.array([2, 1]), np.array([2, 0]))
plt.show()
```

### **Norm of a Vector using Python**

```python
def vec_norm_hand(vecA):
    # instead of this you can use a one liner with numpy: return np.sqrt(np.sum(np.square(vecA)))
    norm = 0
    for i in vecA:
        norm += i**2
    return norm**0.5


vecA = np.array([1, 2])
```

```python
print(f"Vector A: {vecA}")
print(f"Hand    : {vec_norm_hand(vecA)}")
print(f"Scipy (is using Numpy`s (np.linalg.norm) function inside): {la.norm(vecA)}")
print(f"Numpy   : {np.linalg.norm(vecA)}")
print(f"Sklearn has no norm function. It has instead a normalize function.")
plot_2d_vectors(vecA, legend_labels=["Vector A"], title=f"Norm of Vector A: {round(vec_norm_hand(vecA), 3)}")
plt.show()
```

### **Normalization of a Vector using Python**

```python
def vec_normalized_hand(vec):
    normalized_vec = []
    norm = vec_norm_hand(vec)
    for i in vec:
        normalized_vec.append(i/norm)
    return np.array(normalized_vec)


def vec_normalized_numpy(vec):
    return vec / np.linalg.norm(vec)


vecA = np.array([1, 2])
```

```python
print(f"Vector A: {vecA}") 
print(f"Hand    : {vec_normalized_hand(vecA)}, New Norm: {vec_norm_hand(vec_normalized_hand(vecA))}")
print(f"Numpy   : {vec_normalized_numpy(vecA)}, New Norm: {np.linalg.norm(vec_normalized_numpy(vecA))}")
print(f"Sklearn : {normalize([vecA])[0]}, New Norm: {np.linalg.norm(normalize([vecA]).reshape(2))}")
plot_2d_vectors(vecA, normalize([vecA])[0], colormap='Paired', legend_labels=['Original', 'Normalized'], title="Original vs. Normalized Vectors")
```

### **Dot Product using Python**

```python
def vec_dot_hand(vecA, vecB):
    dot = 0
    for a, b in zip(vecA, vecB):
        dot += a * b
    return dot


vecA = np.array([2, 4])
vecB = np.array([5, 3])
vecC = np.array([2, 3])
vecD = np.array([-3, -2])
vecE = np.array([-2, 1])
vecA_normalized = normalize([vecA])[0]
vecB_normalized = normalize([vecB])[0]
vecC_normalized = normalize([vecC])[0]
vecD_normalized = normalize([vecD])[0]
vecE_normalized = normalize([vecE])[0]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plot_2d_vectors(vecA, vecB, vecC, vecD, vecE, legend_labels=["A", "B", "C", "D", "E"], colormap="tab20", title="Non-Normalized Vectors")
plt.subplot(1, 2, 2)   
plot_2d_vectors(vecA_normalized, vecB_normalized, vecC_normalized, vecD_normalized, vecE_normalized, legend_labels=["A", "B", "C", "D", "E"], colormap="tab20", title="Normalized Vectors")
plt.show()
```

### **Cosine Similarity using Python**

```python
def vec_cos_sim_hand(vecA, vecB):
    return vec_dot_hand(vecA, vecB) / (vec_norm_hand(vecA) * vec_norm_hand(vecB))


vecA_normalized = normalize(np.array([[2, 2]]))[0]
vecB_normalized = normalize(np.array([[0, 2]]))[0]
vecC_normalized = normalize(np.array([[-2, 2]]))[0]
vecD_normalized = normalize(np.array([[-2, 0]]))[0]
vecE_normalized = normalize(np.array([[-2, -2]]))[0]

print(f"Vector A {vecA_normalized} and Vector B {vecB_normalized}")
print(f"Hand   : {vec_cos_sim_hand(vecA_normalized, vecB_normalized)}")
print(f"Sklearn: {cosine_similarity([vecA_normalized], [vecB_normalized])[0][0]}")
plot_2d_vectors(vecA_normalized, vecB_normalized, vecC_normalized, vecD_normalized, vecE_normalized, colormap="tab20", legend_labels=["A", "B", "C", "D", "E"], title="Normalized Vectors")
plt.show()
```

```python
print("Calculating the cosine similarity between the vectors A and B, C, D, E.")
print(f"Vector A and Vector A (to simulate two identical vectors): {cosine_similarity([vecA_normalized], [vecA_normalized])[0][0]}")
print(f"Vector A and Vector B: {cosine_similarity([vecA_normalized], [vecB_normalized])[0][0]}")
print(f"Vector A and Vector C: {cosine_similarity([vecA_normalized], [vecC_normalized])[0][0]}")
print(f"Vector A and Vector D: {cosine_similarity([vecA_normalized], [vecD_normalized])[0][0]}")
print(f"Vector A and Vector E: {cosine_similarity([vecA_normalized], [vecE_normalized])[0][0]}")
```

### **Cosine Distance using Python**

```python
def vec_cos_dist_hand(vecA, vecB):
    return 1 - vec_cos_sim_hand(vecA, vecB)


vecA_normalized = normalize(np.array([[2, 2]]))[0]
vecB_normalized = normalize(np.array([[0, 2]]))[0]
vecC_normalized = normalize(np.array([[-2, 2]]))[0]
vecD_normalized = normalize(np.array([[-2, 0]]))[0]
vecE_normalized = normalize(np.array([[-2, -2]]))[0]

print(f"Vector A {vecA_normalized} and Vector B {vecB_normalized}")
print(f"Hand   : {vec_cos_dist_hand(vecA_normalized, vecB_normalized)}")
print(f"Sklearn: {cosine_distances([vecA_normalized], [vecB_normalized])[0][0]}")
plot_2d_vectors(vecA_normalized, vecB_normalized, vecC_normalized, vecD_normalized, vecE_normalized, colormap="tab20", legend_labels=["A", "B", "C", "D", "E"], title="Normalized Vectors")
plt.show()
```

```python
print("Calculating the cosine distance between the vectors A and B, C, D, E.")
print(f"Vector A and Vector A (to simulate two identical vectors): {cosine_distances([vecA_normalized], [vecA_normalized])[0][0]}")
print(f"Vector A and Vector B: {cosine_distances([vecA_normalized], [vecB_normalized])[0][0]}")
print(f"Vector A and Vector C: {cosine_distances([vecA_normalized], [vecC_normalized])[0][0]}")
print(f"Vector A and Vector D: {cosine_distances([vecA_normalized], [vecD_normalized])[0][0]}")
print(f"Vector A and Vector E: {cosine_distances([vecA_normalized], [vecE_normalized])[0][0]}")
```

### **Euclidean Distance using Python**

```python
def vec_euc_dist_hand(vecA, vecB):
    tmp = 0
    for a, b in zip(vecA, vecB):
        tmp += (a - b)**2
    return tmp**0.5


def vec_euc_dist_hand_2(vecA, vecB):
    return np.linalg.norm(vecA - vecB)
```

## **Comparing Vector Embeddings**

- We will use the Python library `sentence-transformers`, which provides various pre-trained models for generating sentence and text embeddings.  
- With these embeddings and the similarity measures we have learned, we can compare how similar different sentences or texts are.  
- For an overview of available models and their performance on different tasks, Hugging Face offers a [benchmark leaderboard](https://huggingface.co/spaces/mteb/leaderboard) that ranks models based on their scores.  
- [Hugging Face](https://huggingface.co/) is a platform that provides a wide range of pre-trained models for various NLP tasks.  

### **Import Libraries**

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import dot_score, cos_sim, euclidean_sim
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances, euclidean_distances
import numpy as np
```

- We are using the model [`paraphrase-multilingual-mpnet-base-v2`](https://huggingface.co/sentence-transformers/paraphrase-multilingual-mpnet-base-v2) for comparing the similarity of sentences.
- This model has following features:
    - **Max Sequence Length**:	128
    - **Dimensions**:	768
    - **Normalized Embeddings**:	false
    - **Suitable Score Functions**:	cosine-similarity (util.cos_sim)
    - **Size**:	970 MB
    - **Pooling**:	Mean Pooling
    - **Training Data**:	Multi-lingual model of paraphrase-mpnet-base-v2, extended to 50+ languages.

### **Load Model**

```python
model = SentenceTransformer("paraphrase-multilingual-mpnet-base-v2")
model.to('cuda')

sentenceA = "I like to eat apples."
sentenceB = "I eat pizza."
sentenceC = "You have a nice car."
sentenceD = "I like to eat bananas."
sentenceE = "I like apples."

combA = [sentenceA, sentenceA]  # two identical sentences, should be cosine similarity 1
combB = [sentenceA, sentenceB]  # two not identical, but similar sentences, should be cosine similarity around 0.5
combC = [sentenceA, sentenceC]  # two different sentences, should be cosine similarity < 1
combD = [sentenceA, sentenceD]  # two similar sentences, should be cosine similarity > 0.5
combE = [sentenceA, sentenceE]  # two similar sentences, should be cosine similarity > 0.5
```

### **Explore generated Embeddings**

- Lets create all embeddings

```python
embeddingsA = model.encode(combA)
embeddingsB = model.encode(combB)
embeddingsC = model.encode(combC)
embeddingsD = model.encode(combD)
embeddingsE = model.encode(combE)
```

- Lets explore one embedding

```python
print(f"We have {len(embeddingsA)} Embeddings for {len(combA)} sentences.")
print(f"Each Embedding has {len(embeddingsA[0])} dimensions.")
print(f"Norm of the first sentence: {np.linalg.norm(embeddingsA[0])}")
print(f"Lowest Value for Example in embeddingsA first sentence: {np.min(embeddingsA[0])}")
print(f"Highest Value for Example in embeddingsA first sentence: {np.max(embeddingsA[0])}")
```

- We can see:
    - **Input Sentence**: "I like to eat apples."
    - **Embedding Dimension**: 768
    - **Norm**: 3.3 (Not Normalized)
    - **Lowest Value**: -0.4707
    - **Highest Value**: 0.4192
- As we know from the Math Fundamentals, the lenght of the vector doesn't matter for cosine similarity, we can use this metric. Otherwise we can normalize the vector and use the dot product or euclidean distance.
- In the following code, we will use all 3 metrics to compare the similarity of the sentences. We will normalize the vectors for the dot product and euclidean distance.

### **Compare Similarity of Sentences**

```python
print(f"Cosine Similarity combA (Sentence Transformer): {cos_sim(*embeddingsA)[0][0]}")
print(f"Cosine Similarity combA (Sklearn)             : {cosine_similarity(embeddingsA)[0][1]}")
print(f"Cosine Similarity combB (Sentence Transformer): {cos_sim(*embeddingsB)[0][0]}")
print(f"Cosine Similarity combB (Sklearn)             : {cosine_similarity(embeddingsB)[0][1]}")
print(f"Cosine Similarity combC (Sentence Transformer): {cos_sim(*embeddingsC)[0][0]}")
print(f"Cosine Similarity combC (Sklearn)             : {cosine_similarity(embeddingsC)[0][1]}")
print(f"Cosine Similarity combD (Sentence Transformer): {cos_sim(*embeddingsD)[0][0]}")
print(f"Cosine Similarity combD (Sklearn)             : {cosine_similarity(embeddingsD)[0][1]}")
print(f"Cosine Similarity combE (Sentence Transformer): {cos_sim(*embeddingsE)[0][0]}")
print(f"Cosine Similarity combE (Sklearn)             : {cosine_similarity(embeddingsE)[0][1]}")
```

- If we are not using normalized vectors, the dot product are calculating a scalar that is difficult to interpret, how similar the sentences to each other are. The length of the vectors has a huge impact on the dot product and the angle between the vectors. 
- So the angle can be high  (should result in a low similarity) but the length of the vectors can be high as well, which results in a high dot product.
- Its better for this metric to use **normalized vectors**.

```python
print(f"Dot Product combA (Sentence Transformer): {dot_score(*embeddingsA)[0][0]}")
print(f"Dot Product combA (Numpy)               : {np.dot(embeddingsA[0], embeddingsA[1])}")
print(f"Dot Product combB (Sentence Transformer): {dot_score(*embeddingsB)[0][0]}")
print(f"Dot Product combB (Numpy)               : {np.dot(embeddingsB[0], embeddingsB[1])}")
print(f"Dot Product combC (Sentence Transformer): {dot_score(*embeddingsC)[0][0]}")
print(f"Dot Product combC (Numpy)               : {np.dot(embeddingsC[0], embeddingsC[1])}")
```

**Lets normalize the vectors and comparing them using the dot product and euclidean distance.**

- First of all normalize the vectors.

```python
combA_normalized = normalize(embeddingsA)
combB_normalized = normalize(embeddingsB)
combC_normalized = normalize(embeddingsC)

print(f"New Norm of the first sentence: {np.linalg.norm(combA_normalized[0])}")
```

- New Norm is 1.0, so the normalization was successful.
- Now lets use the dot product and euclidean distance to compare the similarity of the sentences.

```python
print(f"Dot Product combA (Sentence Transformer): {dot_score(*combA_normalized)[0][0]}")
print(f"Dot Product combA (Numpy)               : {np.dot(combA_normalized[0], combA_normalized[1])}")
print(f"Dot Product combB (Sentence Transformer): {dot_score(*combB_normalized)[0][0]}")
print(f"Dot Product combB (Numpy)               : {np.dot(combB_normalized[0], combB_normalized[1])}")
print(f"Dot Product combC (Sentence Transformer): {dot_score(*combC_normalized)[0][0]}")
print(f"Dot Product combC (Numpy)               : {np.dot(combC_normalized[0], combC_normalized[1])}")
```

```python
print(f"Euclidean Distance combA (Sentence Transformer): {euclidean_sim(*combA_normalized)[0][0]}")
print(f"Euclidean Distance combA (Sklearn)             : {euclidean_distances(combA_normalized)[0][1]}")
print(f"Euclidean Distance combB (Sentence Transformer): {euclidean_sim(*combB_normalized)[0][0]}")
print(f"Euclidean Distance combB (Sklearn)             : {euclidean_distances(combB_normalized)[0][1]}")
print(f"Euclidean Distance combC (Sentence Transformer): {euclidean_sim(*combC_normalized)[0][0]}")
print(f"Euclidean Distance combC (Sklearn)             : {euclidean_distances(combC_normalized)[0][1]}")
```
